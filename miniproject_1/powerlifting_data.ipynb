```python
pwrlft_rows_str = str(source.content) # converting the acquired data to string print(pwrlft_rows_str)

f_out = open("/content/drive/MyDrive/powerlifting/output/crawled_powerlifter_data.csv", "w")
```

```python
with open("/content/drive/MyDrive/powerlifting/output/crawled_powerlifter_data.csv", "w") as f_out:
  start = pwrlft_rows_str.find("rows") + len("rows") + 4
  end  = pwrlft_rows_str.find("const urlprefix") - 9
  print(start, end)



  rows_extr_str = pwrlft_rows_str[start:end]
  rows_extr_str = rows_extr_str.replace("\\n", "")        # extracting only required data from pwrlft_rows_str
  print("Extracted rows: " + rows_extr_str)
  lst_rows = rows_extr_str.split("],[")

  print("list rows:" + str(lst_rows))

  for i in lst_rows:
    cleaned_row = i.strip('[]')
    print(cleaned_row)
    f_out.write(str(cleaned_row) + "\n")

```

```python
with open('/content/drive/MyDrive/powerlifting/output/final_crawled_powerlifter_data.csv', 'w', newline='') as out_f:
    writer = csv.writer(out_f)

    with open('/content/drive/MyDrive/powerlifting/output/crawled_powerlifter_data.csv', newline='') as in_f:
        reader = csv.reader(in_f)

        first_row = next(reader)
        header = ['Index', 'Rank', 'Lifter', 'Username', 'Userid', 'Text_colour', 'Unknown0', 'Unknown1', 'Fed', 'Date', 'Country', 'State', 'unknown3', 'Sex', 'Equip', 'Age', 'Entry', 'Class', 'Weight', 'Squat', 'Bench', 'Deadlift', 'Total', 'Dots']

        writer.writerow(header)
        writer.writerow(first_row)

        for row in reader:
            writer.writerow(row)
```
