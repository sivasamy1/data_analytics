```python
import pandas as pd
from tabulate import tabulate
import numpy as np
```

```python
def pty_print(df):
  print(tabulate(df, headers='keys', tablefmt='github'))
```

## **PART I**

```python
df_world_eco = pd.read_csv('/content/drive/MyDrive/data_analytics/MINIPROJECT_10/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_3731268.csv', skiprows=4).fillna(0)
pty_print(df_world_eco.head())
```

### **1.1 For each country, print Top 10 highest average GDP values accounting only the odd years GDPs.**
### **Expected Output:**
### **country_name, avg_gdp_odd,rank_odd**


```python
def has_odd_number(col):
    try:
        return int(col) % 2 == 1
    except ValueError:
        return False
columns = df_world_eco.columns
odd_cols = [col for col in columns if has_odd_number(col)]
df_odd = df_odd = df_world_eco.loc[:, ["Country Name"] + odd_cols]
df_odd['avg_gdp_odd'] = df_odd[odd_cols].mean(axis=1)
df_odd['rank_odd'] = df_odd['avg_gdp_odd'].rank(ascending=False)
df_odd = df_odd.sort_values(by='rank_odd', ascending=True)
df_odd = df_odd[['Country Name', 'avg_gdp_odd', 'rank_odd']]
pty_print(df_odd.head(11))
```

### **1.2 For each country, print Top 10 highest average GDP values accounting only the even years GDPs.**
### **Expected Output:**
### **country_name, avg_gdp_even,rank_even**


```python
def has_even_number(col):
    try:
        return int(col) % 2 == 1
    except ValueError:
        return False
columns = df_world_eco.columns
even_cols = [col for col in columns if has_even_number(col)]
df_even = df_even = df_world_eco.loc[:, ["Country Name"] + even_cols]
df_even['avg_gdp_even'] = df_even[even_cols].mean(axis=1)
df_even['rank_even'] = df_even['avg_gdp_even'].rank(ascending=False)
df_even = df_even.sort_values(by='rank_even', ascending=True)
df_even = df_even[['Country Name', 'avg_gdp_even', 'rank_even']]
pty_print(df_even.head(11))
```

### **1.3 For each country, print Top 10 highest average GDP values accounting all the years.**
### **Expected Output:**
### **country_name, avg_gdp_all,rank_all**


```python
df_all = df_world_eco.copy()
df_all['avg_gdp_all'] = df_all.drop(columns=df_world_eco.iloc[:, 0:4]).mean(axis=1)
df_all['rank_all'] = df_all['avg_gdp_all'].rank(ascending=False)
df_all = df_all.sort_values(by='rank_all', ascending=True)
df_all = df_all[['Country Name', 'avg_gdp_all', 'rank_all']]
pty_print(df_all.head(11))
```

### **1.4 Merge 1-3 above.**
### **Expected Output:**
### **country_name, avg_gdp_odd, rank_odd, avg_gdp_even, rank_even,avg_gdp_all,rank_all**


```python
df_merge = df_odd.merge(df_even, on='Country Name')
df_merge = df_merge.merge(df_all, on='Country Name')
pty_print(df_merge.head(11))
```

### **Bar 1 -> avg_gdp_odd**
### **Bar 2 -> avg_gdp_even**
### **Bar 3 -> avg_gdp_all**


```python
import matplotlib.pyplot as plt

df = df_merge.iloc[1:]
df = df.head(5)
barWidth = 0.25

r1 = np.arange(len(df['Country Name']))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.bar(r1, df['avg_gdp_odd'], color='#7f6d5f', width=barWidth, edgecolor='white', label='avg_gdp_odd')
plt.bar(r2, df['avg_gdp_even'], color='#557f2d', width=barWidth, edgecolor='white', label='avg_gdp_even')
plt.bar(r3, df['avg_gdp_all'], color='#2d7f5e', width=barWidth, edgecolor='white', label='avg_gdp_all')

plt.xlabel('Country Name')
plt.ylabel('Average GDP')
plt.xticks([r + barWidth for r in range(len(df['Country Name']))], df['Country Name'], rotation=90)

plt.legend()

plt.show()
```

### **2.1 Credit Card Fraud Detection: Credit Card Fraud classification using MLP Classifier. Column value to be classified is in column "Is_fraud" and all other columns are the features. Use two-fold cross validation and evaluation metrics: Precision, Recall, F1-score, Accuracy [wiki]. Also print confusion matrix [wiki]. Refer basic notebook template for some code and also internet**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix

chunk_size = 1000
data_chunks = pd.read_csv("/content/drive/MyDrive/data_analytics/MINIPROJECT_7/fraud test.csv", chunksize=chunk_size)

metrics = {"precision": [], "recall": [], "f1_score": [], "accuracy": [], "conf_matrix": []}
kf = KFold(n_splits=2, shuffle=True, random_state=42)

for chunk in data_chunks:
    chunk = pd.get_dummies(chunk, dtype='int')
    X = chunk.drop(columns=["is_fraud"])
    y = chunk["is_fraud"]

    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)

        y_pred = clf.predict(X_test)

        metrics["precision"].append(precision_score(y_test, y_pred, zero_division=0))
        metrics["recall"].append(recall_score(y_test, y_pred, zero_division=0))
        metrics["f1_score"].append(f1_score(y_test, y_pred, zero_division=0))
        metrics["accuracy"].append(accuracy_score(y_test, y_pred))
        metrics["conf_matrix"].append(confusion_matrix(y_test, y_pred))
```

```python
print("Average Precision:", np.mean(metrics["precision"]))
print("Average Recall:", np.mean(metrics["recall"]))
print("Average F1-score:", np.mean(metrics["f1_score"]))
print("Average Accuracy:", np.mean(metrics["accuracy"]))
```

```python
for i, cm in enumerate(metrics["conf_matrix"], 1):
    print(f"Confusion Matrix - Fold {i}:\n", cm)
```
